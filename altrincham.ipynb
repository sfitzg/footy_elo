Load the dataset of interest.

```python
from match_data import load_match_history

national_2025_26 = load_match_history('nl_25_26.csv', [ 'Date', 'HomeTeam', 'AwayTeam', 'FTR' ])
national_2025_26
```

Compute the elos for the season.

```python
from elo_history import compute_elo_history
history = compute_elo_history(national_2025_26, '2025-08-01', '2026-08-01')
history
```

Plot this with Altrincham highlighted, and mark the date of manager change.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

MY_TEAM = 'Altrincham'

selected_team = history[history['Team'] == MY_TEAM]

sns.lineplot(
    data=selected_team,
    x='Date',
    y='Elo',
    color='red',
    linewidth=2,
    label=MY_TEAM,
)

other_teams = history[history['Team'] != MY_TEAM]

sns.lineplot(
    data=other_teams,
    x='Date',
    y='Elo',
    units='Team',
    estimator=None,
    color='gray',
    alpha=0.5,
    legend=False,
)

# dummy line solely to get 'Others' into the legend
plt.plot(
    [],
    [],
    color='gray',
    alpha=0.5,
    label='Others',
)

# change of manager line
plt.axvline(
    pd.Timestamp('2025-10-25'),
    linestyle=':',
    color='black',
    label='New manager',
)

ax = plt.gca()
ax.xaxis.set_major_locator(mdates.MonthLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b'))


plt.legend()
plt.ylabel('Elo')
plt.xlabel('Date')
plt.title('Altrincham Elo 2025-26')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
